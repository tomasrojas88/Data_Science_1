```python
df = pd.read_csv(url, sep=';', encoding='latin1') # lee el archivo,

df.head()# Muestra las primeras 5 filas
```

```python
df.isnull().sum()  # Cuenta valores faltantes por columna
```

```python
df["origen_provincia"].fillna("No", inplace=True)##rellena los vacios con "No"
```

```python
df["destino_provincia"].fillna("No", inplace=True)##rellena los vacios con "No"
```

```python
df.isnull().sum()  # Cuenta valores faltantes por columna
```

```python
df.info()      # muestra tipos de datos y valores no nulos
```

```python
# transformar una columna de tipo object (o string) a datetime y darle formato: día/mes/año
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'], format='%d/%m/%Y')
```

```python
df.info()      # muestra tipos de datos y valores no nulos
```

```python
# Extraer el mes y el año (puede ser útil para agrupaciones)
df['NumeroMes'] = df['indice_tiempo'].dt.month
df['Año'] = df['indice_tiempo'].dt.year
```

```python
# Verificar el cambio en el tipo de dato y las nuevas columnas
print("Información del DataFrame después de convertir 'Mes':")
df.info()
```

```python
# KDE plot de Distribución Vuelos por Mes
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='NumeroMes', fill=True)
plt.title('Distribución de Vuelos por mes')
plt.xlabel('Mes')
plt.ylabel('Vuelos')
plt.show()
```
