```python
df = pd.read_csv(url, sep=';', encoding='latin1') # lee el archivo,

df.head()# Muestra las primeras 5 filas
```

```python
df.describe()  # Estadísticas básicas de columnas numéricas
```

```python
df.info()      # muestra tipos de datos y valores no nulos
```

```python
df.isnull().sum()  # Cuenta valores faltantes por columna
```

```python
df["origen_provincia"].fillna("No", inplace=True)##rellena los vacios con "No"
```

```python
df["destino_provincia"].fillna("No", inplace=True)##rellena los vacios con "No"
```

```python
df.isnull().sum()  # Cuenta valores faltantes por columna
```

```python
df.info()      # muestra tipos de datos y valores no nulos
```

```python
# transformar una columna de tipo object (o string) a datetime y darle formato: día/mes/año
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'], format='%d/%m/%Y')
```

```python
df.info()      # muestra tipos de datos y valores no nulos
```

```python
# Extraer el mes y el año (puede ser útil para agrupaciones)
df['NumeroMes'] = df['indice_tiempo'].dt.month
df['Año'] = df['indice_tiempo'].dt.year
```

```python
# Verificar el cambio en el tipo de dato y las nuevas columnas
print("Información del DataFrame después de convertir 'Mes':")
df.info()
```

```python
# KDE plot de Distribución Vuelos por Mes
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='NumeroMes', fill=True)
plt.title('Distribución de Vuelos por mes')
plt.xlabel('Mes')
plt.ylabel('Vuelos')
plt.show()
```

```python
# Calcular el indicador de tasa de ocupación
df["Tasa de Ocupación"] = df["pasajeros"] / df["asientos"]
```

```python
# Agrupar por mes y calcular la media de la tasa de ocupación
tasa_ocupacion_mensual = df.groupby('NumeroMes')['Tasa de Ocupación'].mean()
```

```python
# Calcular la tasa de ocupación promedio por mes
tasa_ocupacion_mensual = df.groupby('NumeroMes')['Tasa de Ocupación'].mean()

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(tasa_ocupacion_mensual.index, tasa_ocupacion_mensual.values, marker='o')
plt.title('Tasa de Ocupación Promedio por Mes')
plt.xlabel('Mes')
plt.ylabel('Tasa de Ocupación')

# Nombres de los meses
month_names = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
available_month_labels = [month_names[i-1] for i in tasa_ocupacion_mensual.index]

plt.xticks(ticks=tasa_ocupacion_mensual.index, labels=available_month_labels)
plt.grid()

# Agregar etiquetas encima de cada punto
for x, y in zip(tasa_ocupacion_mensual.index, tasa_ocupacion_mensual.values):
    plt.text(x, y + 0.005, f"{y:.1f}%", ha='center', fontsize=9)

plt.show()
```

```python
tasa_ocupacion_mensual = (
    df.groupby('NumeroMes')[['pasajeros','asientos']]
      .sum()
      .eval("pasajeros / asientos")
)
```

```python
plt.figure(figsize=(10,6))
plt.bar(tasa_ocupacion_mensual.index, tasa_ocupacion_mensual.values*100, color="skyblue")

plt.title("Tasa de Ocupación (sumatoria pasajeros/asientos) por Mes")
plt.xlabel("Mes")
plt.ylabel("Tasa de Ocupación (%)")
plt.xticks(ticks=tasa_ocupacion_mensual.index, labels=available_month_labels, rotation=45)
plt.ylim(70, 100)
plt.grid(axis="x", linestyle="--", alpha=0.6) # Modified to only show x-axis grid lines

# Etiquetas arriba de cada barra
for x, y in zip(tasa_ocupacion_mensual.index, tasa_ocupacion_mensual.values*100):
    plt.text(x, y+0.5, f"{y:.2f}%", ha="center", fontsize=9)

plt.show()
```
